```python
import os
import json
import math
import time
import random
import warnings
from pathlib import Path
from contextlib import nullcontext
from typing import Dict, Any, Optional, Tuple, List

import numpy as np
import pandas as pd

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

from easydict import EasyDict
from tqdm.auto import tqdm

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, balanced_accuracy_score, f1_score, confusion_matrix

warnings.filterwarnings('ignore')

# CONFIG

DRIVE_ROOT = Path('/content/drive/MyDrive')

# Data.
DRIVE_DATA_ROOT = DRIVE_ROOT / 'data/processed_v10_safe'
LOCAL_DATA_ROOT = Path('/content/processed_v10_safe')

# Prefer local archive
if (LOCAL_DATA_ROOT / 'precomputed').exists() and (LOCAL_DATA_ROOT / 'precomputed').is_dir():
    DATA_ROOT = LOCAL_DATA_ROOT
    print('Using local DATA_ROOT:', DATA_ROOT)
else:
    DATA_ROOT = DRIVE_DATA_ROOT
    print('Using Drive DATA_ROOT:', DATA_ROOT)

METADATA_PATH = DATA_ROOT / 'all_metadata.csv'
PRECOMPUTED_DIR = DATA_ROOT / 'precomputed'

# Phase 1 checkpoint.
PHASE1_CKPT = DRIVE_ROOT / 'motionemo_v10_phase1/checkpoints/best_phase1_v10.pt'

# Output.
OUTPUT_ROOT = DRIVE_ROOT / 'motionemo_v10_phase2'
CHECKPOINT_DIR = OUTPUT_ROOT / 'checkpoints'
EVAL_DIR = OUTPUT_ROOT / 'eval'

for p in [OUTPUT_ROOT, CHECKPOINT_DIR, EVAL_DIR]:
    p.mkdir(parents=True, exist_ok=True)

cfg = EasyDict({
    'audio': {
        'n_mels': 80,
        'max_frames': 500,
    },

    'model': {
        'embedding_dim': 512,
        'num_heads': 8,
        'num_layers': 6,
        'dropout': 0.1,
        'state_hidden': 256,
        'num_state_classes': 3,
    },

    'masking': {
        'mask_prob': 0.12,
        'min_span': 3,
        'max_span': 10,
        'min_mask_frames': 4,
        'mask_value': 0.0,
        'mask_prosody_input': True,
    },

    'cpc': {
        'pred_steps': [3, 5, 10],
        'hidden_dim': 512,
        'temperature': 0.12,
        'max_positions': 1024,
        'detach_targets': True,
    },

    'loss': {
        'lambda_state': 1.0,
        'lambda_cpc': 0.12,
        'lambda_recon': 0.20,
        'lambda_smooth': 0.005,
        'lambda_mel_recon': 1.0,
        'lambda_prosody_recon': 0.5,
    },

    'training': {
        'epochs': 60,
        'batch_size': 16,
        'grad_accum_steps': 4,
        'base_lr': 5e-5,
        'min_lr': 5e-7,
        'weight_decay': 1e-4,
        'warmup_epochs': 3,
        'plateau_patience': 6,
        'plateau_factor': 0.5,
        'early_stop_patience': 14,
        'grad_clip': 1.0,
        'num_workers': 2,
        'device': 'cuda' if torch.cuda.is_available() else 'cpu',
        'use_amp': True,
        'seed': 42,
    },

    'runtime': {
        'resume': True,
        'save_every_epoch': True,
    }
})

device = cfg.training.device
amp_enabled = bool(cfg.training.use_amp and device == 'cuda')

print('Device:', device)
print('AMP:', amp_enabled)
print('Data root:', DATA_ROOT)
print('Metadata:', METADATA_PATH)
print('Precomputed:', PRECOMPUTED_DIR)
print('Phase 1 checkpoint:', PHASE1_CKPT)
print('Output:', OUTPUT_ROOT)


# SEED + AMP CONTEXT

def set_seed(seed: int = 42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

set_seed(cfg.training.seed)


def autocast_context():
    if amp_enabled and device == 'cuda':
        return torch.amp.autocast('cuda', enabled=True)
    return nullcontext()



# LOAD METADATA

if not METADATA_PATH.exists():
    raise FileNotFoundError(f'Metadata not found: {METADATA_PATH}')

if not PHASE1_CKPT.exists():
    raise FileNotFoundError(f'Phase 1 checkpoint not found: {PHASE1_CKPT}')

df = pd.read_csv(METADATA_PATH)

required_cols = [
    'sample_id', 'feature_id', 'dataset', 'dataset_id', 'speaker_uid', 'speaker_label',
    'emotion', 'emotion_label', 'split', 'feature_path', 'duration_final', 'lufs_final', 'num_frames'
]
missing = [c for c in required_cols if c not in df.columns]
if missing:
    raise RuntimeError(f'Metadata missing columns: {missing}')

# If using local DATA_ROOT, rewrite feature paths to local precomputed folder.
df['feature_path'] = df['feature_path'].apply(lambda p: str(PRECOMPUTED_DIR / Path(p).name))

sample_path = Path(df.iloc[0]['feature_path'])
print('Sample feature path:', sample_path)
print('Exists:', sample_path.exists())
if not sample_path.exists():
    raise RuntimeError('Sample feature file does not exist. Check local archive extraction or Drive paths.')

print('\nDataset size:', len(df))
print('\nSplit x Dataset')
print(pd.crosstab(df['split'], df['dataset']))
print('\nSplit x Emotion')
print(pd.crosstab(df['split'], df['emotion']))

for a, b in [('train', 'val'), ('train', 'test'), ('val', 'test')]:
    A = set(df[df['split'] == a]['speaker_uid'])
    B = set(df[df['split'] == b]['speaker_uid'])
    overlap = A & B
    print(f'Speaker overlap {a}/{b}: {len(overlap)}')
    if overlap:
        raise RuntimeError(f'Speaker leakage detected: {a}/{b}')

obj = torch.load(sample_path, map_location='cpu', weights_only=False)
print('\nSample keys:', sorted(obj.keys()))
print('Sample mel shape:', obj['mel'].shape)


# VOCAL STATE PROXY FUNCTIONS

STATE_NAMES = ['activation', 'tension', 'stability']
STATE_LABEL_NAMES = ['low', 'mid', 'high']


def _to_np(x):
    if isinstance(x, torch.Tensor):
        return x.detach().cpu().numpy()
    return np.asarray(x)

# tính ba điểm proxy Activation/Tension/Stability từ acoustic descriptors thô của từng mẫu
def compute_vocal_state_scores_from_arrays(
    f0_raw,
    energy_raw,
    zcr_raw,
    voiced_mask,
) -> Dict[str, float]:
    """Compute continuous vocal-state proxy scores from raw frame-level features.

    These scores are not psychological ground-truth. They are weak acoustic
    proxies used to anchor the representation toward vocal dynamics.

    activation: roughly energy/activity/F0 range.
    tension: roughly F0 motion and energy volatility.
    stability: inverse of pause/volatility.
    """
    f0 = _to_np(f0_raw).astype(np.float32).reshape(-1)
    energy = _to_np(energy_raw).astype(np.float32).reshape(-1)
    zcr = _to_np(zcr_raw).astype(np.float32).reshape(-1)
    voiced = _to_np(voiced_mask).astype(np.float32).reshape(-1) > 0.5

    T = min(len(f0), len(energy), len(zcr), len(voiced))
    if T <= 2:
        return {'activation': 0.0, 'tension': 0.0, 'stability': 0.0}

    f0 = f0[:T]
    energy = np.maximum(energy[:T], 0.0)
    zcr = np.maximum(zcr[:T], 0.0)
    voiced = voiced[:T] & np.isfinite(f0) & (f0 > 0)

    eps = 1e-8

    # Robust energy statistics.
    e_mean = float(np.mean(energy))
    e_std = float(np.std(energy))
    e_p95 = float(np.percentile(energy, 95)) + eps
    e_cv = e_std / (e_mean + eps)

    e_rel = energy / e_p95
    pause_ratio = float(np.mean(e_rel < 0.03))

    z_mean = float(np.mean(zcr))
    z_std = float(np.std(zcr))

    if voiced.sum() > 2:
        vf0 = f0[voiced]
        f0_mean = float(np.mean(vf0)) + eps
        f0_p10 = float(np.percentile(vf0, 10))
        f0_p90 = float(np.percentile(vf0, 90))
        f0_range_rel = (f0_p90 - f0_p10) / f0_mean
        f0_diff = np.diff(vf0)
        f0_motion_rel = float(np.std(f0_diff)) / f0_mean
    else:
        f0_range_rel = 0.0
        f0_motion_rel = 0.0

    log_energy = math.log(e_mean + 1e-6)

    activation = 0.45 * log_energy + 0.35 * z_mean + 0.20 * f0_range_rel
    tension = 0.45 * f0_motion_rel + 0.35 * e_cv + 0.20 * z_std
    instability = 0.45 * pause_ratio + 0.35 * f0_motion_rel + 0.20 * e_cv
    stability = -instability

    return {
        'activation': float(activation),
        'tension': float(tension),
        'stability': float(stability),
    }


def assign_state_label(score: float, thresholds: List[float]) -> int:
    lo, hi = thresholds
    if score <= lo:
        return 0
    if score <= hi:
        return 1
    return 2


def compute_or_load_state_thresholds(df_train: pd.DataFrame, force_recompute: bool = False) -> Dict[str, Any]:
    out_path = EVAL_DIR / 'state_thresholds.json'

    if out_path.exists() and not force_recompute:
        with open(out_path, 'r', encoding='utf-8') as f:
            thresholds = json.load(f)
        print('Loaded cached state thresholds:', out_path)
        print(json.dumps(thresholds, indent=2))
        return thresholds

    print('Computing vocal-state thresholds from TRAIN split only...')

    rows = []
    for _, row in tqdm(df_train.iterrows(), total=len(df_train), desc='State thresholds'):
        obj = torch.load(row['feature_path'], map_location='cpu', weights_only=False)
        scores = compute_vocal_state_scores_from_arrays(
            obj['f0_raw'], obj['energy_raw'], obj['zcr_raw'], obj['voiced_mask']
        )
        rows.append({
            'sample_id': row['sample_id'],
            'dataset': row['dataset'],
            **scores,
        })

    score_df = pd.DataFrame(rows)
    score_df.to_csv(EVAL_DIR / 'train_state_proxy_scores.csv', index=False)

    thresholds = {}
    for name in STATE_NAMES:
        q33, q66 = np.percentile(score_df[name].values, [33.333, 66.667])
        thresholds[name] = [float(q33), float(q66)]

    with open(out_path, 'w', encoding='utf-8') as f:
        json.dump(thresholds, f, indent=2, ensure_ascii=False)

    print('Saved thresholds:', out_path)
    print(json.dumps(thresholds, indent=2))

    audit = {}
    for name in STATE_NAMES:
        labels = score_df[name].apply(lambda x: assign_state_label(x, thresholds[name])).values
        audit[name] = pd.Series(labels).value_counts().sort_index().to_dict()
    print('Train state distribution from thresholds:', audit)

    return thresholds


train_df = df[df['split'] == 'train'].reset_index(drop=True)
val_df = df[df['split'] == 'val'].reset_index(drop=True)
test_df = df[df['split'] == 'test'].reset_index(drop=True)

STATE_THRESHOLDS = compute_or_load_state_thresholds(train_df, force_recompute=False)


# CELL 7 — DATASET

class Phase2StateDataset(Dataset):
    def __init__(
        self,
        df_data: pd.DataFrame,
        thresholds: Dict[str, Any],
        max_frames: int = 500,
        deterministic_crop: bool = False,
    ):
        self.df = df_data.reset_index(drop=True)
        self.thresholds = thresholds
        self.max_frames = max_frames
        self.deterministic_crop = deterministic_crop

    def __len__(self):
        return len(self.df)

    def _crop_start(self, T: int) -> int:
        if T <= self.max_frames:
            return 0
        if self.deterministic_crop:
            return max(0, (T - self.max_frames) // 2)
        return random.randint(0, T - self.max_frames)

    def __getitem__(self, idx: int) -> Dict[str, Any]:
        row = self.df.iloc[idx]
        path = Path(row['feature_path'])
        obj = torch.load(path, map_location='cpu', weights_only=False)

        mel = obj['mel'].float()
        f0 = obj['f0'].float()
        energy = obj['energy'].float()
        srate = obj['speaking_rate'].float()

        f0_raw = obj['f0_raw'].float()
        energy_raw = obj['energy_raw'].float()
        zcr_raw = obj['zcr_raw'].float()
        voiced_mask = obj['voiced_mask'].float()

        T = mel.shape[1]
        if T > self.max_frames:
            start = self._crop_start(T)
            end = start + self.max_frames
            mel = mel[:, start:end]
            f0 = f0[start:end]
            energy = energy[start:end]
            srate = srate[start:end]

            f0_raw = f0_raw[start:end]
            energy_raw = energy_raw[start:end]
            zcr_raw = zcr_raw[start:end]
            voiced_mask = voiced_mask[start:end]

        scores = compute_vocal_state_scores_from_arrays(f0_raw, energy_raw, zcr_raw, voiced_mask)
        activation_label = assign_state_label(scores['activation'], self.thresholds['activation'])
        tension_label = assign_state_label(scores['tension'], self.thresholds['tension'])
        stability_label = assign_state_label(scores['stability'], self.thresholds['stability'])

        return {
            'mel': mel,
            'f0': f0,
            'energy': energy,
            'srate': srate,
            'length': mel.shape[1],
            'activation_label': activation_label,
            'tension_label': tension_label,
            'stability_label': stability_label,
            'speaker_id': int(row['speaker_label']),
            'dataset_id': int(row['dataset_id']),
            'emotion_id': int(row['emotion_label']),
            'sample_id': row['sample_id'],
            'dataset': row['dataset'],
        }


def collate_fn(batch):
    max_len = max(item['length'] for item in batch)

    mels, f0s, energies, srates = [], [], [], []
    lengths = []
    act, ten, sta = [], [], []
    spk, dsid, emo = [], [], []
    sample_ids, datasets = [], []

    for item in batch:
        pad = max_len - item['length']
        mels.append(F.pad(item['mel'], (0, pad)).unsqueeze(0))
        f0s.append(F.pad(item['f0'], (0, pad)).unsqueeze(0))
        energies.append(F.pad(item['energy'], (0, pad)).unsqueeze(0))
        srates.append(F.pad(item['srate'], (0, pad)).unsqueeze(0))
        lengths.append(item['length'])

        act.append(item['activation_label'])
        ten.append(item['tension_label'])
        sta.append(item['stability_label'])
        spk.append(item['speaker_id'])
        dsid.append(item['dataset_id'])
        emo.append(item['emotion_id'])
        sample_ids.append(item['sample_id'])
        datasets.append(item['dataset'])

    return {
        'mel': torch.cat(mels, dim=0),
        'f0': torch.cat(f0s, dim=0),
        'energy': torch.cat(energies, dim=0),
        'srate': torch.cat(srates, dim=0),
        'lengths': torch.tensor(lengths, dtype=torch.long),
        'activation_label': torch.tensor(act, dtype=torch.long),
        'tension_label': torch.tensor(ten, dtype=torch.long),
        'stability_label': torch.tensor(sta, dtype=torch.long),
        'speaker_id': torch.tensor(spk, dtype=torch.long),
        'dataset_id': torch.tensor(dsid, dtype=torch.long),
        'emotion_id': torch.tensor(emo, dtype=torch.long),
        'sample_id': sample_ids,
        'dataset': datasets,
    }


# CELL 8 — DATALOADERS + STATE DISTRIBUTION AUDIT

train_ds = Phase2StateDataset(train_df, STATE_THRESHOLDS, max_frames=cfg.audio.max_frames, deterministic_crop=False)
val_ds = Phase2StateDataset(val_df, STATE_THRESHOLDS, max_frames=cfg.audio.max_frames, deterministic_crop=True)

def make_loader_kwargs():
    kwargs = {
        'num_workers': cfg.training.num_workers,
        'pin_memory': (device == 'cuda'),
    }
    if cfg.training.num_workers > 0:
        kwargs.update({'persistent_workers': True, 'prefetch_factor': 2})
    return kwargs

loader_kwargs = make_loader_kwargs()

train_loader = DataLoader(
    train_ds,
    batch_size=cfg.training.batch_size,
    shuffle=True,
    collate_fn=collate_fn,
    **loader_kwargs,
)

val_loader = DataLoader(
    val_ds,
    batch_size=cfg.training.batch_size,
    shuffle=False,
    collate_fn=collate_fn,
    **loader_kwargs,
)

print(f'Train samples: {len(train_ds)} | Val samples: {len(val_ds)}')
print(f'Train batches: {len(train_loader)} | Val batches: {len(val_loader)}')

batch = next(iter(train_loader))
print('Batch mel:', batch['mel'].shape)
print('Batch lengths:', batch['lengths'][:5])
print('Batch activation labels:', batch['activation_label'][:10])


def audit_state_distribution(loader, name='loader', max_batches=None):
    counts = {s: np.zeros(3, dtype=np.int64) for s in STATE_NAMES}
    n = 0
    for i, b in enumerate(tqdm(loader, desc=f'Audit {name} states')):
        for state in STATE_NAMES:
            key = f'{state}_label'
            vals = b[key].numpy()
            for c in range(3):
                counts[state][c] += int((vals == c).sum())
        n += len(b['lengths'])
        if max_batches is not None and (i + 1) >= max_batches:
            break
    print(f'\nState distribution: {name}, n={n}')
    for state in STATE_NAMES:
        print(state, counts[state], counts[state] / max(1, counts[state].sum()))

audit_state_distribution(train_loader, 'train')
audit_state_distribution(val_loader, 'val')



# LENGTH / MASK HELPERS

def make_padding_mask(lengths: torch.Tensor, max_len: int) -> torch.Tensor:
    idx = torch.arange(max_len, device=lengths.device).unsqueeze(0)
    return idx >= lengths.unsqueeze(1)


def make_valid_mask(lengths: torch.Tensor, max_len: int) -> torch.Tensor:
    return ~make_padding_mask(lengths, max_len)


def make_span_mask(lengths: torch.Tensor, max_len: int, cfg_mask: EasyDict, device: str) -> torch.Tensor:
    B = lengths.shape[0]
    mask = torch.zeros(B, max_len, dtype=torch.bool, device=device)

    for b in range(B):
        L = int(lengths[b].item())
        if L <= 0:
            continue

        target = max(cfg_mask.min_mask_frames, int(round(L * cfg_mask.mask_prob)))
        target = min(target, L)
        masked = 0
        attempts = 0
        max_attempts = max(10, target * 3)

        while masked < target and attempts < max_attempts:
            attempts += 1
            span = random.randint(cfg_mask.min_span, cfg_mask.max_span)
            span = min(span, L)
            start = random.randint(0, max(0, L - span))
            end = min(L, start + span)

            before = int(mask[b, :L].sum().item())
            mask[b, start:end] = True
            after = int(mask[b, :L].sum().item())
            masked += max(0, after - before)

        if mask[b, :L].sum() == 0:
            mask[b, random.randint(0, L - 1)] = True

    return mask


def apply_mask_to_inputs(mel, f0, energy, srate, mask, mask_value=0.0, mask_prosody_input=True):
    mel_masked = mel.masked_fill(mask.unsqueeze(1), mask_value)
    if mask_prosody_input:
        f0_masked = f0.masked_fill(mask, 0.0)
        energy_masked = energy.masked_fill(mask, 0.0)
        srate_masked = srate.masked_fill(mask, 0.0)
    else:
        f0_masked, energy_masked, srate_masked = f0, energy, srate
    return mel_masked, f0_masked, energy_masked, srate_masked


def masked_mse_channel_first(pred, target, mask_bt):
    if mask_bt.sum() == 0:
        return torch.tensor(0.0, device=pred.device, requires_grad=True)
    mask = mask_bt.unsqueeze(1).expand_as(pred)
    return F.mse_loss(pred[mask], target[mask])


def masked_mse_channel_last(pred, target, mask_bt):
    if mask_bt.sum() == 0:
        return torch.tensor(0.0, device=pred.device, requires_grad=True)
    mask = mask_bt.unsqueeze(-1).expand_as(pred)
    return F.mse_loss(pred[mask], target[mask])


def temporal_smoothness_loss(z: torch.Tensor, lengths: torch.Tensor) -> torch.Tensor:
    # z: (B, T, D)
    B, T, D = z.shape
    if T <= 1:
        return torch.tensor(0.0, device=z.device, requires_grad=True)

    diff = (z[:, 1:] - z[:, :-1]) ** 2  # (B, T-1, D)
    valid = torch.arange(T - 1, device=z.device).unsqueeze(0) < (lengths - 1).unsqueeze(1)
    if valid.sum() == 0:
        return torch.tensor(0.0, device=z.device, requires_grad=True)
    valid = valid.unsqueeze(-1).expand_as(diff)
    return diff[valid].mean()


# MODEL COMPONENTS

class MotionEmoBackboneV10(nn.Module):
    def __init__(self, n_mels=80, embed_dim=512, num_heads=8, num_layers=6, dropout=0.1):
        super().__init__()

        self.prosody_encoder = nn.Sequential(
            nn.Conv1d(3, 128, kernel_size=5, padding=2),
            nn.BatchNorm1d(128),
            nn.ReLU(),
            nn.Conv1d(128, 256, kernel_size=5, padding=2),
            nn.BatchNorm1d(256),
            nn.ReLU(),
            nn.Conv1d(256, 256, kernel_size=3, padding=1),
            nn.BatchNorm1d(256),
            nn.ReLU(),
        )

        self.mel_encoder = nn.Sequential(
            nn.Conv1d(n_mels, 256, kernel_size=5, padding=2),
            nn.BatchNorm1d(256),
            nn.ReLU(),
            nn.Conv1d(256, 256, kernel_size=3, padding=1),
            nn.BatchNorm1d(256),
            nn.ReLU(),
        )

        self.fusion_proj = nn.Sequential(
            nn.Linear(512, embed_dim),
            nn.LayerNorm(embed_dim),
        )

        enc_layer = nn.TransformerEncoderLayer(
            d_model=embed_dim,
            nhead=num_heads,
            dim_feedforward=embed_dim * 4,
            dropout=dropout,
            batch_first=True,
            norm_first=True,
            activation='gelu',
        )
        self.transformer = nn.TransformerEncoder(enc_layer, num_layers=num_layers)

        self.attn_pool = nn.Sequential(
            nn.Linear(embed_dim, embed_dim // 4),
            nn.Tanh(),
            nn.Linear(embed_dim // 4, 1),
        )

    def forward(self, mel, f0, energy, srate, lengths: Optional[torch.Tensor] = None):
        B, _, T = mel.shape
        prosody_in = torch.stack([f0, energy, srate], dim=1)

        p_feat = self.prosody_encoder(prosody_in)
        m_feat = self.mel_encoder(mel)

        fused = torch.cat([p_feat, m_feat], dim=1)
        x = self.fusion_proj(fused.transpose(1, 2))

        pad_mask = None
        if lengths is not None:
            pad_mask = make_padding_mask(lengths, T)

        H_frame = self.transformer(x, src_key_padding_mask=pad_mask)

        attn_logits = self.attn_pool(H_frame).squeeze(-1)
        if pad_mask is not None:
            # AMP-safe masking.
            attn_logits = attn_logits.masked_fill(pad_mask, -1e4)

        attn = torch.softmax(attn_logits, dim=1).unsqueeze(-1)
        H_utt = (attn * H_frame).sum(dim=1)
        return H_frame, H_utt


class StrongDecoder(nn.Module):
    def __init__(self, embed_dim=512, n_mels=80):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv1d(embed_dim, 512, kernel_size=1),
            nn.BatchNorm1d(512),
            nn.ReLU(),
            nn.Conv1d(512, 512, kernel_size=1),
            nn.BatchNorm1d(512),
            nn.ReLU(),
            nn.Conv1d(512, n_mels, kernel_size=1),
        )

    def forward(self, H_frame):
        return self.net(H_frame.transpose(1, 2))


class ProsodyPredictor(nn.Module):
    def __init__(self, embed_dim=512):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(embed_dim, 256),
            nn.GELU(),
            nn.Linear(256, 3),
        )

    def forward(self, H_frame):
        return self.net(H_frame)


class StateHead(nn.Module):
    def __init__(self, embed_dim=512, hidden_dim=256, num_classes=3):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(embed_dim, hidden_dim),
            nn.LayerNorm(hidden_dim),
            nn.GELU(),
            nn.Dropout(0.1),
            nn.Linear(hidden_dim, num_classes),
        )

    def forward(self, H_utt):
        return self.net(H_utt)


class SampledCPCHead(nn.Module):
    def __init__(self, embed_dim=512, pred_steps=[3, 5, 10], hidden_dim=512, temperature=0.12):
        super().__init__()
        self.pred_steps = pred_steps
        self.temperature = temperature
        self.predictors = nn.ModuleDict({
            str(k): nn.Sequential(
                nn.Linear(embed_dim, hidden_dim),
                nn.GELU(),
                nn.Linear(hidden_dim, embed_dim),
            )
            for k in pred_steps
        })

    def forward(self, z: torch.Tensor, lengths: torch.Tensor, max_positions: int = 1024, detach_targets: bool = True):
        # z: (B, T, D)
        B, T, D = z.shape
        losses = []

        for k in self.pred_steps:
            valid_pairs = []
            for b in range(B):
                L = int(lengths[b].item())
                max_t = L - k
                if max_t > 0:
                    # Pair indices t -> t+k.
                    t_idx = torch.arange(max_t, device=z.device)
                    b_idx = torch.full_like(t_idx, b)
                    valid_pairs.append(torch.stack([b_idx, t_idx], dim=1))

            if not valid_pairs:
                continue

            pairs = torch.cat(valid_pairs, dim=0)  # (N, 2)
            N = pairs.shape[0]
            if N > max_positions:
                perm = torch.randperm(N, device=z.device)[:max_positions]
                pairs = pairs[perm]
                N = max_positions

            b_idx = pairs[:, 0]
            t_idx = pairs[:, 1]

            z_t = z[b_idx, t_idx]
            z_future = z[b_idx, t_idx + k]

            pred = self.predictors[str(k)](z_t)
            target = z_future.detach() if detach_targets else z_future

            pred = F.normalize(pred, dim=-1)
            target = F.normalize(target, dim=-1)

            logits = torch.matmul(pred, target.T) / self.temperature
            logits = torch.clamp(logits, min=-50, max=50)
            labels = torch.arange(N, device=z.device)
            loss = F.cross_entropy(logits, labels)
            losses.append(loss)

        if not losses:
            return torch.tensor(0.0, device=z.device, requires_grad=True)

        return sum(losses) / len(losses)



# INIT MODEL + LOAD PHASE 1

backbone = MotionEmoBackboneV10(
    n_mels=cfg.audio.n_mels,
    embed_dim=cfg.model.embedding_dim,
    num_heads=cfg.model.num_heads,
    num_layers=cfg.model.num_layers,
    dropout=cfg.model.dropout,
).to(device)

decoder = StrongDecoder(
    embed_dim=cfg.model.embedding_dim,
    n_mels=cfg.audio.n_mels,
).to(device)

prosody_pred = ProsodyPredictor(
    embed_dim=cfg.model.embedding_dim,
).to(device)

activation_head = StateHead(cfg.model.embedding_dim, cfg.model.state_hidden, cfg.model.num_state_classes).to(device)
tension_head = StateHead(cfg.model.embedding_dim, cfg.model.state_hidden, cfg.model.num_state_classes).to(device)
stability_head = StateHead(cfg.model.embedding_dim, cfg.model.state_hidden, cfg.model.num_state_classes).to(device)

cpc_head = SampledCPCHead(
    embed_dim=cfg.model.embedding_dim,
    pred_steps=cfg.cpc.pred_steps,
    hidden_dim=cfg.cpc.hidden_dim,
    temperature=cfg.cpc.temperature,
).to(device)

print('Loading Phase 1 checkpoint...')
phase1 = torch.load(PHASE1_CKPT, map_location=device, weights_only=False)
backbone.load_state_dict(phase1['backbone'])
decoder.load_state_dict(phase1['decoder'])
prosody_pred.load_state_dict(phase1['prosody_pred'])
print('Phase 1 loaded successfully.')
print('Phase 1 epoch:', phase1.get('epoch'))
print('Phase 1 best val loss:', phase1.get('best_val_loss'))

params = (
    list(backbone.parameters()) +
    list(decoder.parameters()) +
    list(prosody_pred.parameters()) +
    list(activation_head.parameters()) +
    list(tension_head.parameters()) +
    list(stability_head.parameters()) +
    list(cpc_head.parameters())
)

optimizer = torch.optim.AdamW(
    params,
    lr=cfg.training.base_lr,
    weight_decay=cfg.training.weight_decay,
)

scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer,
    mode='min',
    factor=cfg.training.plateau_factor,
    patience=cfg.training.plateau_patience,
    min_lr=cfg.training.min_lr,
)

scaler = torch.amp.GradScaler('cuda', enabled=amp_enabled)

print('Backbone parameters:', sum(p.numel() for p in backbone.parameters()) / 1e6, 'M')
print('Total parameters:', sum(p.numel() for p in params) / 1e6, 'M')


# LOSS / METRIC HELPERS

def state_loss_and_metrics(logits, labels):
    loss = F.cross_entropy(logits, labels)
    pred = logits.argmax(dim=-1)
    acc = (pred == labels).float().mean()
    return loss, acc.detach(), pred.detach()


def compute_phase2_losses(batch):
    mel = batch['mel'].to(device, non_blocking=True)
    f0 = batch['f0'].to(device, non_blocking=True)
    energy = batch['energy'].to(device, non_blocking=True)
    srate = batch['srate'].to(device, non_blocking=True)
    lengths = batch['lengths'].to(device, non_blocking=True)

    y_act = batch['activation_label'].to(device, non_blocking=True)
    y_ten = batch['tension_label'].to(device, non_blocking=True)
    y_sta = batch['stability_label'].to(device, non_blocking=True)

    B, _, T = mel.shape
    valid_mask = make_valid_mask(lengths, T)
    mask = make_span_mask(lengths, T, cfg.masking, device)

    mel_in, f0_in, energy_in, srate_in = apply_mask_to_inputs(
        mel, f0, energy, srate,
        mask=mask,
        mask_value=cfg.masking.mask_value,
        mask_prosody_input=cfg.masking.mask_prosody_input,
    )

    H_frame, H_utt = backbone(mel_in, f0_in, energy_in, srate_in, lengths=lengths)

    # State heads.
    act_logits = activation_head(H_utt)
    ten_logits = tension_head(H_utt)
    sta_logits = stability_head(H_utt)

    loss_act, acc_act, pred_act = state_loss_and_metrics(act_logits, y_act)
    loss_ten, acc_ten, pred_ten = state_loss_and_metrics(ten_logits, y_ten)
    loss_sta, acc_sta, pred_sta = state_loss_and_metrics(sta_logits, y_sta)

    loss_state = (loss_act + loss_ten + loss_sta) / 3.0

    # CPC.
    loss_cpc = cpc_head(
        H_frame,
        lengths=lengths,
        max_positions=cfg.cpc.max_positions,
        detach_targets=cfg.cpc.detach_targets,
    )

    # Light reconstruction.
    mel_hat = decoder(H_frame)
    prosody_hat = prosody_pred(H_frame)
    prosody_target = torch.stack([f0, energy, srate], dim=-1)

    loss_mel_recon = masked_mse_channel_first(mel_hat, mel, mask)
    loss_pros_recon = masked_mse_channel_last(prosody_hat, prosody_target, mask)
    loss_recon = cfg.loss.lambda_mel_recon * loss_mel_recon + cfg.loss.lambda_prosody_recon * loss_pros_recon

    # Smoothness.
    loss_smooth = temporal_smoothness_loss(H_frame, lengths)

    total_loss = (
        cfg.loss.lambda_state * loss_state +
        cfg.loss.lambda_cpc * loss_cpc +
        cfg.loss.lambda_recon * loss_recon +
        cfg.loss.lambda_smooth * loss_smooth
    )

    metrics = {
        'loss_total': total_loss,
        'loss_state': loss_state.detach(),
        'loss_cpc': loss_cpc.detach(),
        'loss_recon': loss_recon.detach(),
        'loss_smooth': loss_smooth.detach(),
        'loss_act': loss_act.detach(),
        'loss_ten': loss_ten.detach(),
        'loss_sta': loss_sta.detach(),
        'acc_act': acc_act,
        'acc_ten': acc_ten,
        'acc_sta': acc_sta,
        'mask_ratio': (mask.sum().float() / valid_mask.sum().float().clamp_min(1)).detach(),
        'pred_act': pred_act,
        'pred_ten': pred_ten,
        'pred_sta': pred_sta,
        'y_act': y_act.detach(),
        'y_ten': y_ten.detach(),
        'y_sta': y_sta.detach(),
    }

    return total_loss, metrics


def update_metric_sums(sums: Dict[str, float], metrics: Dict[str, Any], batch_size: int):
    scalar_keys = [
        'loss_total', 'loss_state', 'loss_cpc', 'loss_recon', 'loss_smooth',
        'loss_act', 'loss_ten', 'loss_sta', 'acc_act', 'acc_ten', 'acc_sta', 'mask_ratio'
    ]
    for k in scalar_keys:
        sums[k] += float(metrics[k].detach().cpu()) * batch_size
    sums['n'] += batch_size


def finalize_metric_sums(sums: Dict[str, float]) -> Dict[str, float]:
    n = max(1, sums['n'])
    return {k: v / n for k, v in sums.items() if k != 'n'}


def empty_sums():
    return {
        'loss_total': 0.0,
        'loss_state': 0.0,
        'loss_cpc': 0.0,
        'loss_recon': 0.0,
        'loss_smooth': 0.0,
        'loss_act': 0.0,
        'loss_ten': 0.0,
        'loss_sta': 0.0,
        'acc_act': 0.0,
        'acc_ten': 0.0,
        'acc_sta': 0.0,
        'mask_ratio': 0.0,
        'n': 0,
    }


def mean_state_acc(metrics: Dict[str, float]) -> float:
    return float((metrics['acc_act'] + metrics['acc_ten'] + metrics['acc_sta']) / 3.0)


# CHECKPOINT HELPERS

LAST_CKPT = CHECKPOINT_DIR / 'last_phase2.pt'
BEST_LOSS_CKPT = CHECKPOINT_DIR / 'best_phase2_loss.pt'
BEST_STATE_CKPT = CHECKPOINT_DIR / 'best_phase2_state.pt'


def save_checkpoint(path: Path, epoch: int, best_val_loss: float, best_state_acc: float, patience_counter: int, history: Dict[str, Any]):
    torch.save({
        'backbone': backbone.state_dict(),
        'decoder': decoder.state_dict(),
        'prosody_pred': prosody_pred.state_dict(),
        'activation_head': activation_head.state_dict(),
        'tension_head': tension_head.state_dict(),
        'stability_head': stability_head.state_dict(),
        'cpc_head': cpc_head.state_dict(),
        'optimizer': optimizer.state_dict(),
        'scheduler': scheduler.state_dict(),
        'scaler': scaler.state_dict(),
        'cfg': dict(cfg),
        'state_thresholds': STATE_THRESHOLDS,
        'epoch': epoch,
        'best_val_loss': best_val_loss,
        'best_state_acc': best_state_acc,
        'patience_counter': patience_counter,
        'history': history,
        'phase1_ckpt': str(PHASE1_CKPT),
        'data_root': str(DATA_ROOT),
        'metadata_path': str(METADATA_PATH),
    }, path)


def load_checkpoint(path: Path):
    ckpt = torch.load(path, map_location=device, weights_only=False)
    backbone.load_state_dict(ckpt['backbone'])
    decoder.load_state_dict(ckpt['decoder'])
    prosody_pred.load_state_dict(ckpt['prosody_pred'])
    activation_head.load_state_dict(ckpt['activation_head'])
    tension_head.load_state_dict(ckpt['tension_head'])
    stability_head.load_state_dict(ckpt['stability_head'])
    cpc_head.load_state_dict(ckpt['cpc_head'])
    optimizer.load_state_dict(ckpt['optimizer'])
    scheduler.load_state_dict(ckpt['scheduler'])
    if device == 'cuda' and 'scaler' in ckpt:
        scaler.load_state_dict(ckpt['scaler'])
    return ckpt


# RESUME

start_epoch = 1
best_val_loss = float('inf')
best_state_acc = -float('inf')
patience_counter = 0

history = {
    'epoch': [],
    'train_loss': [], 'val_loss': [],
    'train_state': [], 'val_state': [],
    'train_cpc': [], 'val_cpc': [],
    'train_recon': [], 'val_recon': [],
    'train_smooth': [], 'val_smooth': [],
    'train_acc_act': [], 'val_acc_act': [],
    'train_acc_ten': [], 'val_acc_ten': [],
    'train_acc_sta': [], 'val_acc_sta': [],
    'train_acc_mean': [], 'val_acc_mean': [],
    'lr': [],
}

if cfg.runtime.resume and LAST_CKPT.exists():
    print('Resuming from:', LAST_CKPT)
    ckpt = load_checkpoint(LAST_CKPT)
    start_epoch = int(ckpt['epoch']) + 1
    best_val_loss = float(ckpt.get('best_val_loss', best_val_loss))
    best_state_acc = float(ckpt.get('best_state_acc', best_state_acc))
    patience_counter = int(ckpt.get('patience_counter', 0))
    history = ckpt.get('history', history)
    print(f'Resumed at epoch {start_epoch}, best_val_loss={best_val_loss:.6f}, best_state_acc={best_state_acc:.4f}')
else:
    print('Starting Phase 2 training from Phase 1 checkpoint.')



# TRAINING LOOP

for epoch in range(start_epoch, cfg.training.epochs + 1):
    if epoch <= cfg.training.warmup_epochs:
        lr = cfg.training.base_lr * epoch / max(1, cfg.training.warmup_epochs)
        for pg in optimizer.param_groups:
            pg['lr'] = lr

    backbone.train()
    decoder.train()
    prosody_pred.train()
    activation_head.train()
    tension_head.train()
    stability_head.train()
    cpc_head.train()

    optimizer.zero_grad(set_to_none=True)
    train_sums = empty_sums()

    pbar = tqdm(train_loader, desc=f'Phase2 Epoch {epoch:03d}')
    for step, batch in enumerate(pbar):
        with autocast_context():
            loss, metrics = compute_phase2_losses(batch)
            loss_scaled = loss / cfg.training.grad_accum_steps

        scaler.scale(loss_scaled).backward()

        if (step + 1) % cfg.training.grad_accum_steps == 0:
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(params, cfg.training.grad_clip)
            scaler.step(optimizer)
            scaler.update()
            optimizer.zero_grad(set_to_none=True)

        bs = len(batch['lengths'])
        update_metric_sums(train_sums, metrics, bs)

        pbar.set_postfix({
            'loss': f"{float(metrics['loss_total'].detach().cpu()):.3f}",
            'state': f"{float(metrics['loss_state'].cpu()):.3f}",
            'cpc': f"{float(metrics['loss_cpc'].cpu()):.3f}",
            'act': f"{float(metrics['acc_act'].cpu()):.2f}",
            'ten': f"{float(metrics['acc_ten'].cpu()):.2f}",
            'sta': f"{float(metrics['acc_sta'].cpu()):.2f}",
        })

    if len(train_loader) % cfg.training.grad_accum_steps != 0:
        scaler.unscale_(optimizer)
        torch.nn.utils.clip_grad_norm_(params, cfg.training.grad_clip)
        scaler.step(optimizer)
        scaler.update()
        optimizer.zero_grad(set_to_none=True)

    train_metrics = finalize_metric_sums(train_sums)
    train_acc_mean = mean_state_acc(train_metrics)

    # Validation
    backbone.eval()
    decoder.eval()
    prosody_pred.eval()
    activation_head.eval()
    tension_head.eval()
    stability_head.eval()
    cpc_head.eval()

    val_sums = empty_sums()

    with torch.no_grad():
        for batch in tqdm(val_loader, desc=f'Validate {epoch:03d}'):
            with autocast_context():
                loss, metrics = compute_phase2_losses(batch)
            bs = len(batch['lengths'])
            update_metric_sums(val_sums, metrics, bs)

    val_metrics = finalize_metric_sums(val_sums)
    val_acc_mean = mean_state_acc(val_metrics)

    if epoch > cfg.training.warmup_epochs:
        scheduler.step(val_metrics['loss_total'])

    current_lr = optimizer.param_groups[0]['lr']

    print(
        f'Epoch {epoch:03d} | '
        f'Train {train_metrics["loss_total"]:.4f} | Val {val_metrics["loss_total"]:.4f} | '
        f'TrainState {train_metrics["loss_state"]:.4f} | ValState {val_metrics["loss_state"]:.4f} | '
        f'TrainCPC {train_metrics["loss_cpc"]:.4f} | ValCPC {val_metrics["loss_cpc"]:.4f} | '
        f'Act {val_metrics["acc_act"]:.3f} | Ten {val_metrics["acc_ten"]:.3f} | Sta {val_metrics["acc_sta"]:.3f} | '
        f'MeanAcc {val_acc_mean:.3f} | LR {current_lr:.2e}'
    )

    history['epoch'].append(epoch)
    history['train_loss'].append(train_metrics['loss_total'])
    history['val_loss'].append(val_metrics['loss_total'])
    history['train_state'].append(train_metrics['loss_state'])
    history['val_state'].append(val_metrics['loss_state'])
    history['train_cpc'].append(train_metrics['loss_cpc'])
    history['val_cpc'].append(val_metrics['loss_cpc'])
    history['train_recon'].append(train_metrics['loss_recon'])
    history['val_recon'].append(val_metrics['loss_recon'])
    history['train_smooth'].append(train_metrics['loss_smooth'])
    history['val_smooth'].append(val_metrics['loss_smooth'])
    history['train_acc_act'].append(train_metrics['acc_act'])
    history['val_acc_act'].append(val_metrics['acc_act'])
    history['train_acc_ten'].append(train_metrics['acc_ten'])
    history['val_acc_ten'].append(val_metrics['acc_ten'])
    history['train_acc_sta'].append(train_metrics['acc_sta'])
    history['val_acc_sta'].append(val_metrics['acc_sta'])
    history['train_acc_mean'].append(train_acc_mean)
    history['val_acc_mean'].append(val_acc_mean)
    history['lr'].append(current_lr)

    pd.DataFrame(history).to_csv(EVAL_DIR / 'phase2_history.csv', index=False)

    # save last.
    save_checkpoint(
        LAST_CKPT,
        epoch=epoch,
        best_val_loss=best_val_loss,
        best_state_acc=best_state_acc,
        patience_counter=patience_counter,
        history=history,
    )

    # Best by total validation loss.
    improved_loss = val_metrics['loss_total'] < best_val_loss
    if improved_loss:
        best_val_loss = val_metrics['loss_total']
        save_checkpoint(
            BEST_LOSS_CKPT,
            epoch=epoch,
            best_val_loss=best_val_loss,
            best_state_acc=best_state_acc,
            patience_counter=patience_counter,
            history=history,
        )
        print('✓ Best loss checkpoint saved')

    # Best by mean state accuracy.
    improved_state = val_acc_mean > best_state_acc
    if improved_state:
        best_state_acc = val_acc_mean
        save_checkpoint(
            BEST_STATE_CKPT,
            epoch=epoch,
            best_val_loss=best_val_loss,
            best_state_acc=best_state_acc,
            patience_counter=patience_counter,
            history=history,
        )
        print('✓ Best state checkpoint saved')

    if improved_loss or improved_state:
        patience_counter = 0
    else:
        patience_counter += 1
        print(f'No improvement. Patience {patience_counter}/{cfg.training.early_stop_patience}')
        if patience_counter >= cfg.training.early_stop_patience:
            print('Early stopping triggered.')
            break

print('Phase 2 training finished.')
print('Best val loss:', best_val_loss)
print('Best state acc:', best_state_acc)
print('Best loss checkpoint:', BEST_LOSS_CKPT)
print('Best state checkpoint:', BEST_STATE_CKPT)



# PLOT TRAINING HISTORY

hist_path = EVAL_DIR / 'phase2_history.csv'
if hist_path.exists():
    hist = pd.read_csv(hist_path)

    plt.figure(figsize=(9, 5))
    plt.plot(hist['epoch'], hist['train_loss'], label='train_total')
    plt.plot(hist['epoch'], hist['val_loss'], label='val_total')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Phase 2 Total Loss')
    plt.legend()
    plt.tight_layout()
    plt.savefig(EVAL_DIR / 'phase2_total_loss.png', dpi=160)
    plt.show()

    plt.figure(figsize=(9, 5))
    plt.plot(hist['epoch'], hist['train_state'], label='train_state')
    plt.plot(hist['epoch'], hist['val_state'], label='val_state')
    plt.xlabel('Epoch')
    plt.ylabel('State CE')
    plt.title('Phase 2 State Loss')
    plt.legend()
    plt.tight_layout()
    plt.savefig(EVAL_DIR / 'phase2_state_loss.png', dpi=160)
    plt.show()

    plt.figure(figsize=(9, 5))
    plt.plot(hist['epoch'], hist['val_acc_act'], label='activation')
    plt.plot(hist['epoch'], hist['val_acc_ten'], label='tension')
    plt.plot(hist['epoch'], hist['val_acc_sta'], label='stability')
    plt.plot(hist['epoch'], hist['val_acc_mean'], label='mean', linewidth=3)
    plt.xlabel('Epoch')
    plt.ylabel('Validation Accuracy')
    plt.title('Phase 2 Vocal-State Accuracy')
    plt.legend()
    plt.tight_layout()
    plt.savefig(EVAL_DIR / 'phase2_state_accuracy.png', dpi=160)
    plt.show()

    plt.figure(figsize=(9, 5))
    plt.plot(hist['epoch'], hist['train_cpc'], label='train_cpc')
    plt.plot(hist['epoch'], hist['val_cpc'], label='val_cpc')
    plt.xlabel('Epoch')
    plt.ylabel('CPC Loss')
    plt.title('Phase 2 CPC Loss')
    plt.legend()
    plt.tight_layout()
    plt.savefig(EVAL_DIR / 'phase2_cpc_loss.png', dpi=160)
    plt.show()

    print('Saved plots to:', EVAL_DIR)
else:
    print('No history found.')


# QUICK LOAD CHECK

for ckpt_path in [BEST_LOSS_CKPT, BEST_STATE_CKPT, LAST_CKPT]:
    if ckpt_path.exists():
        ckpt = torch.load(ckpt_path, map_location='cpu', weights_only=False)
        print('\nCheckpoint:', ckpt_path.name)
        print('Epoch:', ckpt.get('epoch'))
        print('Best val loss:', ckpt.get('best_val_loss'))
        print('Best state acc:', ckpt.get('best_state_acc'))
    else:
        print('Missing:', ckpt_path)
```
